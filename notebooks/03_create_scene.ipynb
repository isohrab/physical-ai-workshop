# Creating a Custom Scene for LeRobot Arm

## Overview

In this tutorial, we will create a custom simulation scene for the LeRobot arm using NVIDIA Isaac Sim. We'll modify an existing kitchen scene by replacing oranges with a pen object to create a pick-and-place task where the robot places the pen on a plate.

## Initial Scene Setup

The tutorial begins with a kitchen scene containing oranges, as shown below:

![Kitchen with oranges scene](resources/screenshot1.png)

## Step 1: Download Required Assets

1. **Download the scene files:**

   - Navigate to: https://github.com/LightwheelAI/leisaac/releases/tag/v0.1.0
   - Download and extract the scene files to your **workspace** folder

2. **Download the robot model:**
   - Download `s0101_follower.usd`
   - Place it in: `~/workspace/leisaac/assets/robots/`

## Step 2: Open and Modify the Scene

### Launch Isaac Sim and Load Scene

1. Start NVIDIA Isaac Sim
2. Navigate to **File > Open**
3. Go to the workspace folder and select the `kitchen_with_orange` folder then in the right side select the `scene.usd`.

### Remove Orange Objects

1. Locate the **Stage** panel on the right side of the interface
2. Find the orange objects in the stage hierarchy
3. Select and delete all orange objects from the scene

## Step 3: Add Pen Object

### Access NVIDIA Assets Browser

1. Navigate to **Window > Browsers > NVIDIA Assets**
2. A new NVIDIA Assets tab will appear in the bottom panel

### Add Mechanical Pen

1. In the NVIDIA Assets browser, search for "MechanicalPen"

   ![NVIDIA Assets browser with MechanicalPen search](resources/screenshot2.png)

2. Drag and drop the MechanicalPen onto the kitchen counter near the plate
3. Position the pen so it rests properly on the surface
4. Right click on the MechanicalPencil in the stage tree (right panel) then choose `Add > Physics > Rigid body with Colliders preset`

The final scene should appear as follows:

![Kitchen scene with pen positioned on counter](resources/screenshot3.png)

## Step 4: Save the Modified Scene

### Collect and Save Scene

1. Navigate to **File > Collect and Save As**
2. Save the scene as `kitchen_with_pen` in the following directory:
   ```
   /home/ubuntu/workspace/leisaac/assets/scenes/
   ```

finally close the Isaac Sim without saving the current scene
